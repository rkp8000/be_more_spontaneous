#### Stats

Right: The probability that the network follows the most probable sequence starting from each of the two starting nodes. This is higher than chance but not equal to 1.

#### Stats

Left: when we drive the most probable sequence starting from node 0, the probability of the network actually generating that path increases substantially.
Right: Here we consider the network state in which node X was active at the last time step, and in which we drive node Y (which is not anatomically downstream of node X) at the current time step. We plot the probability that node Y activates vs the probability that a node downstream from X activates as a function of the drive amplitude. This shows when drive is low, the network follows its intrinsic patterns, but when the drive is high, the network follows the drive.
