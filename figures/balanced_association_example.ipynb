## Association units selective for pairs of sequentially activated upstream units
An activation-based associative memory network requires units that are selective for two specific upstream units becoming activated nearby in time. A mathematical challenge is creating a unit that activates in response to its two preferred units activating sequentially, but not to one of those units activating twice in a row.

Here I show that including a single inhibitory unit that receives equal projections from all "buffer units" (the units with long time constants that allow for spaced out activation of the upstream units) and which inhibits all buffer units can make said units (and more importantly the association units that are downstream from them) exhibit this type of selectivity.

The fundamental idea is (1) to note that if an upstream unit activates twice, all buffer units connected to it will try to activate, and therefore (2) to add an inhibitory unit that silences all the buffers if more than one buffer is active. This can be done by setting the inputs to the inhibitory unit from the buffers such that one buffer being active keeps it well below its the threshold, whereas several buffers being active makes it leap over its firing threshold and silence everything.
